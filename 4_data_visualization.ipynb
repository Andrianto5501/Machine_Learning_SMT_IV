```python
import pandas as pd
import matplotlib.pyplot as plt

df_btc = pd.read_csv('../coin_Bitcoin.csv') # read csv file
df_btc = df_btc.set_index('Date') #setting attribute Date as index

df_btc.head(100)[['High','Low']].plot.line()
plt.title("Bitcoin Price Changes") # title of Chart
plt.ylabel("Harga") # y axis label
plt.xlabel("Waktu") # x axis label
plt.show()
```

```python
df_btc.head(100)[['High','Low']].plot.area()
plt.title("Bitcoin Price Changes") # title of Chart
plt.ylabel("Harga") # y axis label
plt.xlabel("Waktu") # x axis label
plt.show()
```

```python
df_btc.head(100)[['High','Low']].plot.box()
plt.title("Bitcoin Price Changes") # title of Chart
plt.ylabel("Harga") # y axis label
plt.show()
```

```python
import numpy as np

df_stroke = pd.read_csv('../dataset_stroke.csv')
df_stroke['bmi'] = df_stroke['bmi'].replace(np.nan, df_stroke['bmi'].median())

df_stroke.head(100)[['bmi']].plot.bar()
plt.title("Body Mass Index") # title
plt.ylabel("BMI")
plt.xlabel("Id")
plt.show()
```

```python
df_stroke['age'].plot.hist(bins=20,color='green')
plt.title("Age Range") # title
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
frequency_data = df_stroke['smoking_status'].value_counts()
frequency_data.plot.pie(autopct='%.2f',labels= None,figsize=(6,6)) #autopct to display procentage
plt.title("Smokers Procentage")
plt.legend(df_stroke['smoking_status'],loc=2) # placed legends
plt.show()
```

```python
plt.scatter(df_stroke['age'],df_stroke['bmi'],color='red')
plt.title("Age and BMI")
plt.ylabel('bmi')
plt.xlabel('Age')
plt.show()
```

```python
frequency_data = df_stroke['Residence_type'].value_counts()
frequency_data.plot.pie(autopct='%.2f',labels= None,figsize=(6,6)) #autopct to display procentage
plt.title("Residence Type Procentage")
plt.legend(df_stroke['Residence_type'],loc=2) # placed legends
plt.show()
```
